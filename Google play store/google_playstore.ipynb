```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\data cleaning\\datasets\\Google play store\\googleplaystore.csv")
df
```

```python
df['Reviews'] = df['Reviews'].str.replace('M','*1e6', regex = True).map(pd.eval).astype(float)
df['Reviews']
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
print(df.duplicated().sum())
df
```

```python
# checking the number of apps with rating = 5

df[df['Rating'] == 5]

df['Rating'].value_counts().plot(kind = 'bar')
# a.plot(kind = 'pie')
```
